```python
import os

import pandas as pd
from matplotlib import pyplot as plt
```

```python
models = os.listdir("../result/valid_save")
for model in models:
    result = pd.read_csv("../result/valid_save/" + model + "/result.csv")
    plt.plot(result["score"], label=model)
models = os.listdir("../result/valid")
for model in models:
    result = pd.read_csv("../result/valid/" + model + "/result.csv")
    plt.plot(result["score"], label=model)
plt.legend()
plt.show()
print(models)
```
